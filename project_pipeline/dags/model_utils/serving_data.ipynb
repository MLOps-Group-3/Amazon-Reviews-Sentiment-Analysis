```python
import pandas as pd
import json
import os

# Load your data
csv_file_path = "/home/niresh/Downloads/cleaned_data.csv"  # Replace with your CSV file path

# Read the CSV into a DataFrame
df = pd.read_csv(csv_file_path)
print("CSV loaded successfully!")
```

```python
df.head()
```

```python
# Ensure 'review_year' and 'review_month' columns are present and correctly formatted
if 'year' not in df.columns or 'review_month' not in df.columns:
    raise ValueError("The dataset must contain 'year' and 'review_month' columns.")

# Group the data by 'review_year' and 'review_month'
grouped_data = df.groupby(['year', 'review_month'])
print("Data grouped by year and month.")
```

```python
# Directory to save JSONL files
output_dir = "Serving Batches"
os.makedirs(output_dir, exist_ok=True)

# Loop through each group and save as JSONL
for (year, month), group in grouped_data:
    # Generate the file name
    jsonl_file_path = os.path.join(output_dir, f"{year}_{month}.jsonl")
    
    # Save the group to a JSONL file
    with open(jsonl_file_path, "w") as jsonl_file:
        for record in group.to_dict(orient='records'):
            jsonl_file.write(json.dumps(record) + "\n")
    
    print(f"Saved {jsonl_file_path}.")
```
